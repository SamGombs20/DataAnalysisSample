```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn import datasets, tree
```

```python
iris = datasets.load_iris()
features = iris.data
target = iris.target
features[0], target.shape
```

```python
target
```

```python
decision_tree = DecisionTreeClassifier()
model = decision_tree.fit(features, target)
model
```

```python
observation = [[5,4,3,2]]
model.predict(observation)
```

```python
model.predict_proba(observation)
```

```python
diabetes = datasets.load_diabetes()
features = diabetes.data
target = diabetes.target
features.shape, target[0:10]
```

```python
decision_tree_reg = DecisionTreeRegressor(random_state=0)
model = decision_tree_reg.fit(features, target)
model
```

```python
import pydotplus
from IPython.display import Image
```

```python
dot_data = tree.export_graphviz(decision_tree, feature_names=iris.feature_names, class_names=iris.target_names)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
